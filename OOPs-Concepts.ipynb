## GETTER | SETTER | PROPERTY

```python
class Salary:
    def __init__(self, salary):
        self._salary = salary

    def get_salary(self):
        return self._salary

    def set_salary(self, new_salary):
        if isinstance(new_salary, float) and new_salary > 10:
            self._salary = new_salary
        else:
            print("Invalid Salary")
```

```python
rocky = Salary(1000)
```

```python
rocky._salary
```

```python
rocky.salary
```

```python
rocky.get_salary()
```

```python
rocky.set_salary(0.4)
```

```python
rocky.get_salary()
```

### DECORATOR

```python
# A Python decorator is a function that takes in a function and returns it by adding some functionality
```

```python
class Salary:
    def __init__(self, salary):
        self._salary = salary

    @property
    def salary(self):
        print('Executing getter funtion')
        return self._salary

    @salary.setter
    def salary(self, new_salary):
        print('Executing setter funtion')
        if isinstance(new_salary, float) and new_salary > 10:
            self._salary = new_salary
        else:
            print("Invalid Salary")
```

```python
Rahul = Salary(5000)
```

```python
Rahul.salary
```

```python
Rahul.salary = 7007.077
```

```python
print('updated - ',Rahul.salary)
```

### Accessing class object from another class

```python
class StudentAndCollegeRelation:
    
    def __init__(self, nameOfCollege, nameOfStudent):
        self.nameOfCollege = nameOfCollege
        self.nameOfStudent = nameOfStudent

    def printStudentAndCollege(self):
        print(self.nameOfStudent, ' studying in ', self.nameOfCollege)

class Student:
    def __init__(self, studentName, sandc):
        self.studentName = studentName
        self.sandc = sandc
    def printStudentAndCollege(self):
        self.sandc.printStudentAndCollege()
```

```python
Dimat = StudentAndCollegeRelation('DIMAT', 'Rocky')
```

```python
StuObj = Student('Rocky', Dimat)
```

```python
StuObj.printStudentAndCollege()
```

## METHOD OVERLOADING

2 methods in  the same class have the same name but different parameters, so when we call the method, the version that is executed
is determined by the data types of the arguments or by the number of arguments

```python
class BepecCT:
    def __init__(self):
        self.topics = []
        
    def add_lectures(self, session):
        print(f"Adding {session.capitalize()} lecture to the dashboard")
        self.topics.append(session)
        print(f"{session.capitalize()} was added into the dashboard")

class KanthCT(BepecCT):
    def add_lectures(self, session):
        print("Notifications!  Your dashboard got unlocked with New lecture")
        BepecCT.add_lectures(self,session)
        print("List of sessions: ", self.topics)
```

```python
dslectures = KanthCT()
```

```python
dslectures.add_lectures("Node.js")
```

## POLYMORPHISM

```python
class File:
    def __init__(self, name, extension):
        self.name = name
        self.extension = extension
    def open(self):
        print("Opening a regular file")

class PDFFile(File):
    def __init__(self, name):
        super().__init__( name, '.pdf')
    def open(self):
        print("Opening pdf file!")

class TEXTFile(File):
    def __init__(self, name):
        super().__init__( name, '.txt')
    def open(self):
        print("Opening text file!")
```

```python
pdf = PDFFile("OOPS")
```

```python
pdf.open()
```

```python
txt = TEXTFile("JS")
```

```python
txt.open()
```
