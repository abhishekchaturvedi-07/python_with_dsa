### Concatenation of Data Frames

```python
df1 = pd.DataFrame(
    {
        "A":["A0","A1","A2","A3"],
        "B":["B0","B1","B2","B3"],
        "C":["C0","C1","C2","C3"],
        "D":["D0","D1","D2","D3"],
    },
    index = [0,1,2,3],
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "A":["A4","A5","A6","A7"],
        "B":["B4","B5","B6","B7"],
        "C":["C4","C5","C6","C7"],
        "D":["D4","D5","D6","D7"],
    },
    index = [4,5,6,7],
)
df2
```

```python
df3 = pd.DataFrame(
    {
        "A":["A8","A9","A10","A11"],
        "B":["B8","B9","B10","B11"],
        "C":["C8","C9","C10","C11"],
        "D":["D8","D9","D10","D11"],
    },
    index = [8,9,10,11]
)
df3
```

```python
frames = [df1,df2,df3]
concatOperation = pd.concat(frames)
concatOperation
```

```python
sumAxis0 = concatOperation.sum(axis=0)
```

```python
sumAxis0
```

```python
sumAxis1 = concatOperation.sum(axis=1)
sumAxis1
```

```python
# Approach 2 


df2 = {'Name': 'Raj', 'Maths': 90, 'Science': 93}
df = df._append(df2, ignore_index = True)
display(df)
```

```python
# Approach 3 

dict = {'Name':['Rocky', 'Rahul', 'Raj', 'Georgia'],
        'Maths':[98, 91, 97, 95],
        'Science':[95, 99, 84, 76]
       }

df1 = pd.DataFrame(dict)
print('---------- Dict 1 to DF 1 -----------')
display(df1)

dict = {'Name':['Amit', 'Hruthik'],
        'Maths':[80, 94],
        'Science':[73, 82]
       }
df2 = pd.DataFrame(dict)
print('---------- Dict 2 to DF 2 -----------')
display(df2)

df3 = pd.concat([df1, df2], ignore_index = True)
df3.reset_index()

print('---------- Concatenated DF 3 -----------')
display(df3)
```

### DROP Rows and Columns with Null values

```python
newdf
```

```python
# Drop with Rows having null value

droppedRowxWithNull = newdf.dropna(axis=0)
display(droppedRowxWithNull)
```

```python
# Drop with Columns having null value

droppedColumnsWithNull = newdf.dropna(axis=1)
display(droppedColumnsWithNull)
```

```python
# SUBSET - based on specific subset of a column - like if there will be null in maths then only delete, and not checking science
```

```python
droppedSubsetWithNull = newdf.dropna(subset= [ 'Maths'])
display(droppedSubsetWithNull)
```
